```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("car_crashes")
```

```python
df.head()
```

```python
# State
# total - Number of drivers involved in fatal collisions per billion miles	National Highway Traffic Safety Administration, 2012
# speeding Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding	National Highway Traffic Safety Administration, 2009
# alcohol Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired	National Highway Traffic Safety Administration, 2012
# not_distracted Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted	National Highway Traffic Safety Administration, 2012
# no_previous Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents	National Highway Traffic Safety Administration, 2012
# ins_permium Car Insurance Premiums ($)	National Association of Insurance Commissioners, 2011
# ins_losses Losses incurred by insurance companies for collisions per insured driver ($)	National Association of Insurance Commissioners, 2010
```

```python
df_texas = df[df["abbrev"] == "TX"]
df_alabama = df[df["abbrev"] == "AL"]
df_florida = df[df["abbrev"] == "FL"]
```

```python
estados = ["Texas", "Alabama", "Florida"]
```

```python
pct_acidentes_alcool = [df_texas["alcohol"].iloc[0], df_alabama["alcohol"].iloc[0], df_florida["alcohol"].iloc[0]]
pct_acidentes_alcool
```

```python
# Gráfico de Barras
plt.figure(figsize=(6, 4))
plt.bar(estados, pct_acidentes_alcool, color=["green", "blue", "purple"])
plt.title("(%) dos acidentes fatais que tinham excesso de álcool (por estado)")
plt.xlabel("Estado")
plt.ylabel("(%) Álcool")
plt.tight_layout()
plt.show()
```

```python
df_filtered = df.query("abbrev in ['TX', 'AL', 'FL']")
df_filtered
```

```python
df_filtered.plot(kind="bar" ,x="abbrev", y="alcohol")

plt.title("(%) dos acidentes fatais que tinham excesso de álcool (por estado)")

plt.xlabel("Estado")
plt.ylabel("(%) Álcool")

plt.show()
```
