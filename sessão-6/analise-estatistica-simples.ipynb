```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("anscombe")
```

```python
df.head(3)
```

```python
sns.lmplot(data=df, x="x", y="y", col="dataset", col_wrap=2)
plt.show()
```

```python
df_I = df[df["dataset"] == "I"]
df_II = df[df["dataset"] == "II"]
df_III = df[df["dataset"] == "III"]
df_IV = df[df["dataset"] == "IV"]
```

```python
print(df_I["y"].mean(), df_II["y"].mean() , df_III["y"].mean() , df_IV["y"].mean())
```

```python
print(df_I["y"].median(), df_II["y"].median() , df_III["y"].median() , df_IV["y"].median())
```
