```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")

# Segredinho - se colocarem o tema do seaborn, ele afeta também o matplotlib
sns.set_theme(style='whitegrid')
```

```python
df.head(3)
```

```python
df.dtypes
```

```python
# Histograma - Analisar Distribuições (de vendas, gorjetas, etc...)
plt.hist(df["total_bill"], bins=10)
plt.axvline(df["total_bill"].mean(), color="green")
plt.axvline(df["total_bill"].median(), color="purple")
```

```python
# Gráfico de Barras - Variável Categórica + Variável Numérica (somatório, média, máximos/mínimos, total, count)

count_per_day = df["day"].value_counts()
count_per_day = count_per_day.reindex(["Thur", "Fri", "Sat", "Sun"])

plt.bar(count_per_day.index, count_per_day.values, color=["lightblue", "#bff", "pink", "lightgreen"])
```

```python
# Gráfico de Dispersão - Relacionar Duas Variáveis Numéricas
plt.scatter(df["total_bill"], df["tip"])
```

```python
# Histograma - Agora com o Seaborn
# Aqui temos KDE - uma curva suavizada que estima a forma da distribuição (tipo um histograma contínuo)
sns.histplot(data=df, x="total_bill", kde=True)
plt.show()
```

```python
# boxplot - Analisar Distribuição em função de categorias
# Analisar a distribuição das gorjetas ao almoço vs jantar
sns.boxplot(data=df, x="time", y="tip")
```

```python
# Gráfico de Dispersão com Linha de Regressão Linear
sns.lmplot(data=df, x="total_bill", y="tip")
```

```python
# Heatmap - Comparar colunas numéricas
# Vamos primeiro selecionar as colunas numéricas
df_numericas = df[["size", "tip", "total_bill"]]
sns.heatmap(data=df_numericas.corr())
```
