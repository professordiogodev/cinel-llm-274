```python
df_smokers = df[df["smoker"] == "Yes"]
df_non_smokers = df[df["smoker"] == "No"]
```

```python
# Desafio:
# Quem é que dá mais tips? quem fuma ou quem não fuma?
print("Tip média dos fumadores:", df_smokers["tip"].mean())
print("Tip média dos não fumadores:", df_non_smokers["tip"].mean())
```

# Transformação

```python
# Coluna que mostra a percentagem da gorjeta
df["tip_pct"] = (df["tip"] / df["total_bill"] * 100).round(2)
```
