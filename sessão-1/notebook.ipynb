# O Meu Primeiro Notebook
Vamos explorar listas e ciclos em Python.

```python
# Criar uma lista de temperaturas (°C)
temperaturas = [22, 25, 19, 30, 28, 15, 21]
print(f"Temperaturas: {temperaturas}")
print(f"Número de registos: {len(temperaturas)}")
```

```python
# Calcular a média
media = sum(temperaturas) / len(temperaturas)
print(f"Temperatura média: {media:.1f}°C")
```

```python
# Encontrar máxima e mínima
print(f"Máxima: {max(temperaturas)}°C")
print(f"Mínima: {min(temperaturas)}°C")
```

```python
# Filtrar dias quentes (acima da média)
dias_quentes = [t for t in temperaturas if t > media]
print(f"Dias acima da média: {dias_quentes}")
print(f"Quantidade: {len(dias_quentes)} de {len(temperaturas)} dias")
```
