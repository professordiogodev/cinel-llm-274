# Análise de Turma
Vamos analisar a média da turma, e listar quem aprovou e quem reprovou.

```python
turma = {
  "Ana": 15,
  "Bruno": 8,
  "Carla": 18,
  "Diogo": 12,
  "Eva": 9
}
```

```python
notas = []
```

```python
for aluno in turma:
    nota = turma[aluno]
    notas.append(nota)

    if (nota >= 10):
        print(f"O aluno { aluno } passou.")
    else:
        print(f"O aluno { aluno } não passou.")

media = sum(notas) / len(notas)
print(f"Média da turma: {media}")
```

# Conclusões
A média da turma positiva, no entanto, dois dos alunos não passaram.
